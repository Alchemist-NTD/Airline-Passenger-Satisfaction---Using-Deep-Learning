```python
cleaned_train_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AI Project/cleaned_train.csv')
```

Các cột rating cần chuẩn hóa về trong khoảng [0,1] : Inflight wifi service, Departure/Arrival time convenient,  Ease of Online booking, Food and drink, Online boarding, Seat Comfort, Inflight entertainment, On-board service, Leg room service, Baggage handling, Checkin service, Inflight service, Cleanliness

```python
def standardRating(df, cols):
  # S is a series
  for col in cols:
    m = np.max(df[col])
    df[col] = df[col].apply(lambda x: x / m)
    
  return df
```

```python
cleaned_train_df = standardRating(cleaned_train_df, ['Inflight wifi service', 'Departure/Arrival time convenient', 
                                                     'Ease of Online booking', 'Food and drink', 'Online boarding',
                                                     'Seat comfort', 'Inflight entertainment', 'On-board service', 
                                                     'Leg room service', 'Baggage handling', 'Checkin service', 
                                                     'Inflight service', 'Cleanliness'])
```
